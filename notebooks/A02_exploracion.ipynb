## Analisis de texto

```python
import nltk
nltk.download('book')
from nltk.book import *
from nltk.corpus import stopwords
```

```python
print(stopwords.words('spanish'))
```

```python
textos_compras = ''
for row in compras_df['CONCEPTO']:
    textos_compras = textos_compras + ' ' + row
```

```python
stopwords_list = stopwords.words('spanish') + ['2020', 'cont','millar','dpp','est','op','ret','mty']
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wc = WordCloud(background_color='white')
wc.generate(textos_compras)

plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
```

```python
textos_compras = ' '.join([x for x in textos_compras.strip().lower().split(' ') if x not in stopwords_list and len(x)>4])
```

```python
wc = WordCloud(background_color='white')
wc.generate(textos_compras)

plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
```
